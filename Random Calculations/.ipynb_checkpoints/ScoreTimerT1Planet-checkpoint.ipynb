```python
from sympy import *
```

```python
f = Function('f')
x, y = symbols('x y')
c0, c1, c2, c3, c4, c5, c6, c7, c8 = symbols('c0, c1, c2, c3, c4, c5, c6, c7, c8', constants=True)

eq = Eq(f(x), c0 + c1*x + c2*x**2 + c3*x**3 + c4*x**4 + c5*x**5)
eq
```

```python
eq_diff = Eq(diff(eq.lhs, x), diff(eq.rhs, x))
eq_diff
```

```python
min_max = solveset(eq_diff.rhs, x, domain=S.Reals)
```

```python
min_max.args[1].args[0]
```
